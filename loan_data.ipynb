```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("loan_data.csv")
```

```python
print("Original Shape:", df.shape)
print(df.head())
```

Check Null Values

```python
print("\nNull values in each column:")
print(df.isnull().sum())
```

 Remove rows with null values

```python
df = df.dropna()
```

```python
print("\nShape after removing null values:", df.shape)
```

 Remove Duplicate Data

```python
duplicates = df.duplicated().sum()
print("\nDuplicate rows:", duplicates)
```

```python

df = df.drop_duplicates()
```

```python

print("Shape after removing duplicates:", df.shape)
```

 Basic Data Cleaning

```python
# Convert age to integer (if needed)
df["person_age"] = df["person_age"].astype(int)
```

```python
# Remove unrealistic ages
df = df[df["person_age"] > 18]
```

```python
print("\nCleaned data shape:", df.shape)
```

Visualization (Graphs)

```python
# Graph 1 â€” Income Distribution
plt.figure()
plt.hist(df["person_income"], bins=30)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# Graph 2 â€” Loan Amount vs Credit Score
plt.figure()
plt.scatter(df["loan_amnt"], df["credit_score"])
plt.title("Loan Amount vs Credit Score")
plt.xlabel("Loan Amount")
plt.ylabel("Credit Score")
plt.show()
```

```python
# Graph 3 â€” Gender Count
plt.figure()
df["person_gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
